```python
from langgraph.store.memory import InMemoryStore
from typing import Dict

# Initialize store
store = InMemoryStore()

# Define namespace and memories
namespace = ("user_id", "memories")

memories = [
    {
        "id": "mem1",
        "data": {
            "food_preference": "I like pizza and pasta",
            "animal_preference": "I like dogs and cats",
            "hobby": "I enjoy reading books"
        }
    },
    {
        "id": "mem2",
        "data": {
            "food_preference": "I love sushi",
            "animal_preference": "I have two parrots",
            "hobby": "I play guitar"
        }
    }
]

# Store memories
for memory in memories:
    store.put(namespace, memory["id"], memory["data"])

# Query and format results
def format_query_results(query_results, query_type):
    print(f"\n{query_type} Results:")
    print("-" * 50)
    for item in query_results:
        print(f"Memory ID: {item.key}")
        print(f"Content: {item.value[query_type.lower().replace(' ','_')+'_preference']}")
        print("-" * 50)

# Execute queries with formatting
food_query = store.search(namespace, query="What foods does the user like?")
format_query_results(food_query, "Food")
```
